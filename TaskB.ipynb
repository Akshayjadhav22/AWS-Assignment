# Task B – AWS Glue ETL (Spark)
# Run this code is Glue ETL Spark session

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum

spark = SparkSession.builder.appName('SmartGearETL').getOrCreate()

df = spark.read.option('header', 'true').option('inferSchema','true').csv(
    's3://smartgear-ak-rajan-dev/raw/smartgear_sales.csv'
)

region_qty = df.groupBy('Region').agg(
    sum('Quantity').alias('total_quantity')
)

region_qty.show()

region_qty.write.mode('overwrite').parquet(
    's3://smartgear-ak-rajan-dev/curated/region_qty/'
)
```
